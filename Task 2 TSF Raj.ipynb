# "Spliting The Data".

```python
# splitting the data into 2 parts x and y.
# x=data[independent variables].
# y=data[dependent variables].
x=data.iloc[:,[0,1,2,3]].values
y=data['Species']
```

```python
y.head()
```

```python
# unique values in y.
y.unique()
```

# Applying Kmeans Algorithm.

```python
# finding the optimum number of clusters from kmeans classification using the elbow method.
from sklearn.cluster import KMeans
elbow=[]
for i in range(1,9):
    kmeans=KMeans(n_clusters=i, random_state=0)
    kmeans.fit(x)
    elbow.append(kmeans.inertia_)
    
plt.figure(figsize=(8,4))
plt.plot(range(1,9),elbow,'ro-',color='red')
plt.title('The Elbow Method',fontsize=15,color='blue')
plt.xlabel('Number of clusters',fontsize=15,color='red')
plt.ylabel('Within cluster sum of squares(WCSS)',fontsize=15,color='black')
plt.grid()
plt.show()
```

##### <font color = 'red'>The above graph shows 3 as the optimal number of clusters . After 3 , the line becomes linear in nature. </font>

# "Creating the kmeans Classifier to predict".

```python
# creating the kmeans classifier.
kmeans=KMeans(n_clusters=3)
y_pred=kmeans.fit_predict(x)
```

# "predicting the model".

```python
y_pred
```

# "Visualization after applying the Kmeans Algorithm".

```python
# visualization of PetalLengthCm vs PetalWidthCm.
plt.figure(figsize=(10,6))
plt.scatter(x[y_pred == 0,2],x[y_pred == 0,3],label='Iris-setosa', color='blue')
plt.scatter(x[y_pred == 1,2],x[y_pred == 1,3],label='Iris-versicolour',color = 'orange')
plt.scatter(x[y_pred == 2,2],x[y_pred == 2,3],label='Iris-virginica',color = 'green')
plt.scatter(kmeans.cluster_centers_[:,2],kmeans.cluster_centers_[:,3],label = 'Centroids',color ='red')
plt.title('PetalLengthCm vs PetalWidthCm',fontsize=15,color = 'blue')
plt.xlabel('PetalLengthCm',fontsize=15, color = 'red')
plt.ylabel('PetalWidthCm',fontsize=15)
plt.legend()
plt.show()
```
